# Employee Appraisal Regression Analysis
This notebook explores a dataset with employee appraisal data and builds a Linear Regression model.

**Features used**:
- Rating
- Behavior
- Experience
- Band (encoded as Band_Num)

**Target**:
- Appraisal

- Spot linear or nonlinear relationships

### Data Preparation

```python
X = df[['Rating', 'Behavior', 'Experience', 'Band_Num']]
y = df['Appraisal']

X2 = df[['Rating','Experience']]
X3 = df[['Rating','Behavior','Experience']]
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y, test_size=0.25, random_state=0)
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y, test_size=0.3, random_state=0)
```

### Train Linear Regression Model

```python
model2 = LinearRegression()
model2.fit(X2_train, y2_train)
y2_pred = model2.predict(X2_test)

print("R2 Score:", r2_score(y2_test, y2_pred))
print("MSE:", mean_squared_error(y2_test, y2_pred))
print("Coefficients:", model2.coef_)
print("Intercept:", model2.intercept_)
```

```python
model3 = LinearRegression()
model3.fit(X3_train, y3_train)
y3_pred = model3.predict(X3_test)

print("R2 Score:", r2_score(y3_test, y3_pred))
print("MSE:", mean_squared_error(y3_test, y3_pred))
print("Coefficients:", model3.coef_)
print("Intercept:", model3.intercept_)
```

## Linear equation for Salary Prediction

For 2 variables -
Appraisal = 20.2 + 10.2(Rating) + 1.5(Experience)

For 3 variables -
Appraisal = 2.8 + 10.2(Rating) + 5.0(Behavior) + 1.5(Experience)


### K-Fold Cross-Validation

K-Fold Cross Validation is a technique used to evaluate the performance of a machine learning model by splitting the dataset into K equal parts (folds):

- The model is trained on K-1 folds and validated on the remaining fold.
- This process is repeated K times, each time using a different fold as the validation set.
- The final performance is the average of all K validation scores.
###  Why use it?
- It helps reduce overfitting.
- Provides a more robust estimate of model performance.
- Ensures that every data point is used for both training and validation.



```python
# For 2 variables
kf = KFold(n_splits=4, shuffle=True, random_state=1)
scores2 = cross_val_score(model2, X2, y, cv=kf, scoring='r2')
print('2 Variables')
print("K-Fold R2 Scores:", scores2)
print("Average R2 Score:", np.mean(scores2))

# For 3 variables
kf = KFold(n_splits=4, shuffle=True, random_state=1)
scores3 = cross_val_score(model3, X3, y, cv=kf, scoring='r2')
print('3 Variables')
print("K-Fold R2 Scores:", scores3)
print("Average R2 Score:", np.mean(scores3))
```

```python
print('2 Variables')
# Visualizing R2 score variation across folds
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(scores2)+1), scores2, marker='o', linestyle='--', color='blue')
plt.title("R² Score per Fold")
plt.xlabel("Fold")
plt.ylabel("R² Score")
plt.ylim(0, 1)
plt.grid(True)
plt.xticks(range(1, len(scores2)+1))
plt.show()

# Alternatively, use barplot
plt.figure(figsize=(8, 5))
sns.barplot(x=list(range(1, len(scores2)+1)), y=scores2, palette='viridis')
plt.title("R² Score Variation Across Folds")
plt.xlabel("Fold")
plt.ylabel("R² Score")
plt.ylim(0, 1)
plt.show()
```
